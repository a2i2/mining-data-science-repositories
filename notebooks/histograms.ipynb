```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("../data/boa/github_api.csv", parse_dates=['created_at', 'updated_at', 'pushed_at'])
```

```python
ds.number_of_contributors.value_counts().sort_index()
```

```python
nonds = pd.read_csv("../data/non-data-science/github_api.csv", parse_dates=['created_at', 'updated_at', 'pushed_at'])
```

```python
def plot_all(dim):
    xs1 = ds[dim]
    xs2 = nonds[dim]
    start = max(1, min(np.min(xs1), np.min(xs2)))
    end = max(1, np.max(xs1), np.max(xs2))
    logbins = np.logspace(np.log10(start),np.log10(end),30)
    
    plt.hist(xs1, bins=logbins)
    plt.xscale('log')
    plt.title('distribution of ' + dim + ' for DS projects')
    plt.xlabel(dim)
    plt.ylabel('count')
    plt.savefig(dim + '-ds.png', dpi=300)
    plt.show()
    
    plt.hist(xs2, bins=logbins)
    plt.xscale('log')
    plt.title('distribution of ' + dim + ' for Non-DS projects')
    plt.xlabel(dim)
    plt.ylabel('count')
    plt.savefig(dim + '-nonds.png', dpi=300)
    plt.show()
```

```python
dims = ['stars', 'size', 'forks_count', 'open_issues_count', 'watch', 'number_of_contributors']
```

```python
for dim in dims:
    plot_all(dim)
```

```python
def plot_all_date(dim):
    xs1 = ds[dim]
    xs2 = nonds[dim]
    start = min(np.min(xs1), np.min(xs2))
    end = max(np.max(xs1), np.max(xs2))
    
    plt.hist(xs1, bins=30, range=(start, end))
    plt.title('distribution of ' + dim + ' for DS projects')
    plt.xlabel(dim)
    plt.ylabel('count')
    plt.savefig(dim + '-ds.png', dpi=300)
    plt.show()
    
    plt.hist(xs2, bins=30, range=(start, end))
    plt.title('distribution of ' + dim + ' for Non-DS projects')
    plt.xlabel(dim)
    plt.ylabel('count')
    plt.savefig(dim + '-nonds.png', dpi=300)
    plt.show()
```

```python
dims_date = ['created_at', 'updated_at', 'pushed_at']
```

```python
for dim in dims_date:
    plot_all_date(dim)
```
